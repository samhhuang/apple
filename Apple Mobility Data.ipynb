```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:\\Users\\Downloads\\applemobilitytrends.csv')
df
```

```python
region = ['Germany','India','Indonesia','Japan','Malaysia','Philippines','Singapore','Thailand','United States']
```

```python
df2 = df.loc[df['region'].isin(region),:]
df2
```

```python
df3 = df2.dropna(axis='columns')
df3 = df3.iloc[:,1:]
df3
```

```python
df4 = np.transpose(df3)
df4
```

```python
df5 = df4.iloc[2:,:]
df5
```

```python
df5.columns = ['Germany,driving','Germany,transit','Germany,walking',
               'India,driving','India,walking',
               'Indonesia,driving','Indonesia,walking',
               'Japan,driving','Japan,transit','Japan,walking',
               'Malaysia,driving','Malaysia,walking',
               'Philippines,driving','Philippines,transit','Philippines,walking',
               'Singapore,driving','Singapore,transit','Singapore,walking',
               'Thailand,driving','Thailand,walking',
               'United States,driving','United States,transit','United States,walking']
df5
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3)
df5[['United States,driving','United States,transit','United States,walking']].plot(ax=ax[0],figsize=(18,4))
df5[['Japan,driving','Japan,transit','Japan,walking']].plot(ax=ax[1])
df5[['Germany,driving','Germany,transit','Germany,walking']].plot(ax=ax[2])
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3)
df5[['Philippines,driving','Philippines,transit','Philippines,walking']].plot(ax=ax[0],figsize=(18,4))
df5[['Singapore,driving','Singapore,transit','Singapore,walking']].plot(ax=ax[1])
df5[['Thailand,driving','Thailand,walking']].plot(ax=ax[2])
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3)
df5[['India,driving','India,walking']].plot(ax=ax[0],figsize=(18,4))
df5[['Indonesia,driving','Indonesia,walking']].plot(ax=ax[1])
df5[['Malaysia,driving','Malaysia,walking']].plot(ax=ax[2])
```
